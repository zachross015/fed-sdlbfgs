```python
import pandas as pd
import glob

files = glob.glob('outputs*')

for file in files:
    print(file)
```

```python
dfs = []
for file in files:
    df = pd.read_csv(file)
    info = file.split('_')
    df['model'] = info[1]
    df['optim'] = info[2]
    df = df[(df['epoch'] <= 500)]
    dfs.append(df)
df = pd.concat(dfs)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

model = 'CNN'

fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ys = ['tr_acc', 'tr_loss', 'te_acc', 'te_loss']

for y, axs in zip(ys, ax.flatten()):
    sns.lineplot(df[df.model == model], x='epoch', y=y, hue='optim', ax=axs)

fig.suptitle(f'Performance analysis on {model}')
handles, labels = ax[1][1].get_legend_handles_labels()
fig.legend(handles, labels)

for axs in ax.flatten():
    axs.legend([])

plt.show()
```

- SdlBFGS eventually matches training accuracy of Adam and eventually surpasses all testing accuracies (albeit it takes a while)
- SdLBFGS Layer is able to outperform regular SdLBFGS, but has much worse runtime
- KFAC (only the first 6 epochs are shown) shows promising results, but has convergence issues
- Shampoo is also shown but runtime is very slow

```python
axs = sns.barplot(df[df.model == model], x='optim', y='elapsed')
axs.set_ylabel('mean time elapsed per epoch (s)')
```

```python
axs = sns.barplot(df[(df.model == model) & (df.optim != 'Shampoo') & (df.optim != 'SdLBFGSLayer')], x='optim', y='elapsed')
axs.set_ylabel('mean time elapsed per epoch (s)')
```
